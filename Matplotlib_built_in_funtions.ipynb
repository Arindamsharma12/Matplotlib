<h1>Matplotlib</h1>

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Types of Plots in this library**

<ol>
    <li>Linear Plot</li>
    <li>Bar Plot</li>
    <li>Histogram</li>
    <li>Pie Plot</li>
    <li>Scattered</li>
</ol>

<h2>1. Linear Plot</h2>

```python
x = np.array([1,2,3,4])
y = np.array([1,2,3,4])
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Linear Plot")
plt.show()
```

<h2>2. Bar Plot</h2>

```python
category = ['Physics','Chemistry','Maths']
value = [90,34,87]

bar = plt.bar(category,value,color=['green','orange','red'])
plt.xlabel("Highest Marks")
plt.ylabel("Subject")
plt.title("Bar Plot")
plt.show()
```

<h2>3. Histogram</h2>

```python
import random
import numpy as np
```

```python
number = np.random.randint(10,60,(50))
print(number)
```

```python
hist = plt.hist(number,edgecolor='black',color='grey',label= "python")
plt.title("Histogram")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# plt.grid()
plt.legend()
```

<h2>4. Pie Plot</h2>

```python
x = [10,20,30,40]
y = ["C++","Java","Python","Javascript"]
c = ['r','b','g','y']
```

```python
pie = plt.pie(x,labels = y,colors = c,autopct = "%0.1f%%")
plt.title("Pie Plot")
plt.show()
```
